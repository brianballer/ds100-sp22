## GROUP BY operations

```python
%%sql
SELECT *
FROM Dish;
```

```python
%%sql
SELECT type
FROM Dish;
```

```python
%%sql
SELECT type
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, MAX(cost)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, 
       SUM(cost), 
       MIN(cost),
       MAX(name)
FROM Dish
GROUP BY type;
```

## Trickier GROUP BY

```python
%%sql
SELECT type, COUNT(cost)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, COUNT(cost)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, COUNT(*)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, cost
FROM Dish
GROUP BY type, cost;
```

```python
%%sql
SELECT type, cost, COUNT(*)
FROM Dish
GROUP BY type, cost;
```

```python
%%sql
SELECT type, COUNT(*)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, COUNT(*)
FROM Dish
GROUP BY type
HAVING MAX(cost) < 8
```

```python
%%sql
SELECT type, COUNT(*)
FROM Dish
WHERE cost < 8
GROUP BY type;
```

## DISTINCT

```python
%%sql
SELECT DISTINCT cost
FROM Dish;
```

```python
%%sql
SELECT DISTINCT type 
FROM Dish
WHERE cost < 9;
```

```python
%%sql
SELECT DISTINCT type, cost 
FROM Dish
WHERE cost < 9;
```

```python
%%sql
SELECT DISTINCT type, cost 
FROM Dish;
```

```python
%%sql
SELECT type, AVG(cost)
FROM Dish
GROUP BY type;
```

```python
%%sql
SELECT type, AVG(DISTINCT cost)
FROM Dish
GROUP BY type;
```
